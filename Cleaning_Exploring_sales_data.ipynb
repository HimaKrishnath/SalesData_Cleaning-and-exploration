```python
#Another way of adding 


#let us use .apply()
def get_state(y):
    return y.split(',')[1]
def get_c(y):
    return y.split(',')[2].split(' ')[1]


df['State']=df["Purchase Address"].apply(lambda x: f'{get_state(x)} ({get_c(x)})')
df.head()
```

```python
res=df.groupby('State').sum()
res.head()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Hour']=df['Order Date'].dt.hour
df['Minute']=df['Order Date'].dt.minute
df.head()
```

```python
#What products are most often sold together?
```

```python
df_dul=df[df['Order ID'].duplicated(keep=False)]
df_dul['Grouped']=df_dul.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df_dul=df_dul[['Order ID','Grouped']].drop_duplicates()
df_dul.head()
```

```python
from itertools import combinations
from collections import Counter

count=Counter()
for row in df_dul['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
count.most_common(10)

for key, value in count.most_common(10):
    print(key, value)
```
